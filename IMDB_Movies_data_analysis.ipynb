**lets do some joins**

```python
#let us create a dataframe which consists of only english movies
english_movies = imdb_df[imdb_df['original_language']=='en']
```

```python
#let us create a dataframe which consists of long movies
long_movies = imdb_df[imdb_df['runtime']>=120]['imdb_id']
```

```python
english_movies.shape,long_movies.shape
```

```python
#merging the two dataframes using inner join
merged_ij_df = pd.merge(english_movies,long_movies,how='inner',left_on='imdb_id',right_on='imdb_id')
```

```python
merged_ij_df
```

```python
temp_df = imdb_df[(imdb_df['original_language']=='en')&(imdb_df['runtime']>=100)]
```

```python
temp_df
```

```python
#merging two datafarmes by using left join
merged_lj_df = pd.merge(english_movies,long_movies,how='left',left_on='imdb_id',right_on='imdb_id')
```

```python
merged_lj_df
```

```python
#merging two datafarmes using right join
merged_rj_df = pd.merge(english_movies,long_movies,how='right',left_on='imdb_id',right_on='imdb_id')
```

```python
merged_rj_df
```

```python
#merging dataframes using outer join
merged_oj_df = pd.merge(english_movies,long_movies,on='imdb_id',how='outer')
```

```python
merged_oj_df
```

**some groupby operations**

```python
 #import datetime modules
 from datetime import datetime
 from datetime import date
```

```python
imdb_df['new_release_date']= imdb_df['release_date'].apply(lambda x : datetime.strptime(x,'%m/%d/%y'))
```

```python
imdb_df['year_of_release']=imdb_df['new_release_date'].apply(lambda x : x.year)
```

```python
imdb_df
```

```python
imdb_df['year_of_release'].unique()
```

```python
imdb_df['year_of_release']=np.where(imdb_df['year_of_release']>2020,imdb_df['year_of_release'] - 100,imdb_df['year_of_release'])
```

```python
#we can find maximum movie runtime across different years
max_runtime = imdb_df.groupby('year_of_release')['runtime'].max().reset_index()
```

```python
max_runtime
```

```python
max_runtime.rename(columns={'runtime':'max_runtime'},inplace = True)
```

```python
max_runtime
```

```python
imdb_df.groupby('year_of_release')['imdb_id'].count().reset_index()
```

```python
imdb_df.groupby('year_of_release',as_index = False).agg(({"runtime":"sum"}))
```

```python
#group the dataframe by month and item and extract a number of stats fromm each group
imdb_df.groupby(['year_of_release'],as_index=False).agg({'runtime':"mean",'revenue':"sum",'imdb_id':"count"})\
.rename(columns={'runtime':'avg_runtime','revenue':'total_revenue','imdb_id':'num_movies'})
```
